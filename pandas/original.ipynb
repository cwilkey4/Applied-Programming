```python
import pandas as pd
```

```python
app = pd.read_csv("practice/pandas/googleplaystore.csv", index_col=0)
app.head()
```

```python
app_filtered = app.dropna()
```

```python
category_installations = app_filtered.groupby('Category')['App'].count() 

highest_categories = category_installations.sort_values(ascending=False)

top_ten = highest_categories.head(25)

print(top_ten)
```
